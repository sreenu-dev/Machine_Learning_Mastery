Map and Apply(axis=column,index)

```python
price_mean=house_data.Price.mean()
house_data.Price.map(lambda x: x-price_mean)
```

```python
def priceMean(row):
    row.Price=row.Price-price_mean
    return row

house_data.apply(priceMean,axis='columns')
```
